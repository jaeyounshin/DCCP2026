```python
words = ["바나나", "사과", "체리", "딸기"]
print(sorted(words, key=len))
def last_char(s: str) -> str:
    return s[-1]
print(sorted(words, key=last_char))
```

```python
def add(*numbers):
    total = 0
    for n in numbers:
        total += n
    return total

print(add(10, 20, 30))
```

```python
def greet(name:str, greeting: str = "안녕하세요") -> str:
    """인사말을 반환하는 함수"""
    return f"{greeting}, {name}!"
print(greet("철수"))
print(greet("영희", "반감습니다"))
```
